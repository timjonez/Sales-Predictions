```python
import pandas as pd
import numpy as np
%matplotlib inline
```

 Read in the dataset, set date as index

```python
df1 = pd.read_csv('bullets_sales.csv',index_col='Date',parse_dates=True)
```

```python
df1.head()
```

filter out all 'TF' rows

```python
df1 = df1[df1['Channel']=='RS']
```

```python
df1.head()
```

```python
df1.index
```

Resample data into daily frequency

```python
df1 = df1.resample(rule='D').sum()
```

```python
df1['Quantity'].plot()
```

```python
from statsmodels.tsa.stattools import adfuller
```

Check non-stationarity of data

```python
adfuller(df1['Quantity'])
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

Evaluate seasonality aspect of the data

```python
result = seasonal_decompose(df1['Quantity'])
```

```python
result.plot()
```

Find optimal P,D,Q values for ARIMA/SARIMA model 

```python
from pmdarima import auto_arima
```

```python
auto_arima(df1['Quantity']).summary()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

Split data into train/test and fit model to training set

```python
train = df1.iloc[:-365]
test = df1.iloc[-365:]
model = SARIMAX(train['Quantity'],order=(2, 1, 3))
```

```python
result = model.fit()
```

Predict values for test data

```python
start = len(train)
end = len(train)+len(test)-1
prediction = result.predict(start,end).rename('Predictions')
```

```python
prediction.plot(figsize=(15,8))
```

Compare predictions to real data

```python
test['Quantity'].plot(figsize=(15,8),legend=True)
prediction.plot()
```

```python
train.plot(figsize=(15,8))
prediction.plot(legend=True)
```

```python
from statsmodels.tools.eval_measures import rmse
```

Calculate root mean squared error to measure model performance

```python
rmse(test['Quantity'],prediction)
```
